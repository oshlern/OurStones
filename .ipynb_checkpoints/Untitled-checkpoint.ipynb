```python
import pickle
import numpy as np
import matplotlib.pyplot as plt


# f.pickle is balancing, g.pickle is not balancing
with open("f.pickle", "rb") as f:
    data_balancing = pickle.load(f)
    
# Here the puck slid away from the robot meaning that the robot needs to update its
# positioning by moving the tcp closer to the base (+y)
with open("g.pickle", "rb") as f:
    data_not_balancing = pickle.load(f)
```

```python
times = np.array([t for t, _ in data_balancing])
wrenches = np.array([ft for _, ft in data_balancing])
len(times)
```

```python
start = 9000

# Forces plot
plt.plot(times[start:], wrenches[start:,:3])
plt.show()

# Torques plot
plt.plot(times[start:], wrenches[start:,3:])
plt.show()
```

```python
times = np.array([t for t, _ in data_not_balancing])
wrenches = np.array([ft for _, ft in data_not_balancing])
len(times)
```

```python
start = 9000

# Forces plot
plt.plot(times[start:], wrenches[start:,:3])
plt.show()

# Torques plot
plt.plot(times[start:], wrenches[start:,3:])
plt.show()
```

```python
F_u = wrenches[-1,:3]
T_u = wrenches[-1,3:]
```

```python
np.linalg.norm(T_u) / np.linalg.norm(F_u) # 0.025 is set a balancing threshold
```
